```python
import requests
import pandas as pd
import sqlite3
```

```python
url = "https://api.themoviedb.org/3/discover/movie"
params = {
    "api_key": API_KEY,
    "language": "en-US",
    "sort_by": "popularity.desc",
    "page": 1
}
response = requests.get(url, params=params)
data = response.json()
movies = data["results"]
len(movies)
```

```python
df = pd.DataFrame(movies)
df = df[[
    "id",
    "title",
    "release_date",
    "popularity",
    "vote_average",
    "vote_count",
    "genre_ids"
]]
df.head()
```

```python
import sqlite3
df["genre_ids"] = df["genre_ids"].astype(str)
conn = sqlite3.connect("movies.db")
df.to_sql("movies", conn, if_exists="replace", index=False)
print("Data stored in SQLite database")
```

```python
movies_df = pd.read_sql("SELECT * FROM movies", conn)
movies_df.head()
```
